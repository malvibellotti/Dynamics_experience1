# Visualizzazione dati

```python
f = open("figure8.out", "r")
lines = f.readlines()
print(len(lines))

elements= []
for line in lines:
    elements = np.append(elements, [line.split()])
print (len(elements))

n_bodies = int(lines[0])
n_timelapses = int(len(lines) / (2+n_bodies))
print(n_timelapses)

'''
event = 0 
print(elements[ event:(event + 2 + (n_bodies*1) +(n_bodies*6)) ])
event = event + 2 +(n_bodies*1) + n_bodies*6
print(event)
print(elements[event :(event + 2 + (n_bodies*1) +(n_bodies*6)) ])
'''
```

```python
timelapse = [["n_bodies","time", "1", "x", "y", "z", "vx", "vy", "vz"]]

#print(timelapse)
   
event = 0


for i in range(n_timelapses):
    array = elements[event :(event + 2 + (n_bodies*1) +(n_bodies*6))]
    #print(array)
    timelapse.append(array)
    event = event + 2 + (n_bodies*1) + n_bodies*6

#print(timelapse)
#print (timelapse[0])
print (timelapse[3][3])
#print (timelapse[2])

```

```python
#qua raga ho salvato tutti i valori, sicuramente lo posso fare meglio con un ciclo, con tre corpi ancora è ok, di più no ahaha
x1 = timelapse[1][3] 
y1 = timelapse[1][4]
z1 = timelapse[1][5]
vx_1 = timelapse[1][6]
vy_1 = timelapse[1][7]
vz_1 = timelapse[1][8]

x2 = timelapse[1][10]
y2 = timelapse[1][11]
z2 = timelapse[1][12]
vx_2 = timelapse[1][13]
vy_2 = timelapse[1][14]
vz_2 = timelapse[1][15]

x3 = timelapse[1][16]
y3 = timelapse[1][17]
z3 = timelapse[1][18]
vx_3 = timelapse[1][19]
vy_3 = timelapse[1][20]
vz_3 = timelapse[1][21]

for i in range(n_timelapses-1):
    x1 = np.append(x1, timelapse[i+2][3])
    y1 = np.append(y1, timelapse[i+2][4])
    z1 = np.append(z1,timelapse[i+2][5])
    vx_1 = np.append(vx_1,timelapse[i+2][6])
    vy_1 = np.append(vy_1,timelapse[i+2][7])
    vz_1 = np.append(vz_1,timelapse[i+2][8])
                       
    x2 = np.append(x2,timelapse[i+2][10])
    y2 = np.append(y2, timelapse[i+2][11])
    z2 = np.append(z2, timelapse[i+2][12])
    vx_2 = np.append(vx_2, timelapse[i+2][13])
    vy_2 = np.append(vy_2, timelapse[i+2][14])
    vz_2 = np.append(vz_2, timelapse[i+2][15])

    x3 = np.append(x3,timelapse[i+2][16])
    y3 = np.append(y3, timelapse[i+2][17])
    z3 = np.append(z3, timelapse[i+2][18])
    vx_3 = np.append(vx_3, timelapse[i+2][19])
    vy_3 = np.append(vy_3, timelapse[i+2][20])
    vz_3 = np.append(vz_3, timelapse[i+2][21])

print(x1, y1)
```

## plotting the position

```python
plt.figure(figsize=(10, 8))
#IF I WANT POINTS:

for i in range(len(x1)):
    plt.scatter(x1[i], y1[i])
for i in range(len(x2)):
    plt.scatter(x2[i], y2[i])
for i in range(len(x3)):
    plt.scatter(x3[i], y3[i])


#IF I WANT THE PLOT OF THE ORBITS
plt.plot(x1, y1, label='1', color = 'red')
plt.plot(x2, y2, label='2', color = 'green')
plt.plot(x3, y3, label='3', color = 'blue')

plt.title('three masses simulation...')
plt.xlabel('x')
plt.ylabel('y')
#plt.gca().set_aspect('equal', adjustable='box')
plt.grid(True)
plt.legend()
plt.show()
```

## plotting the velocity

```python
plt.figure(figsize=(10, 8))
#IF I WANT POINTS:

for i in range(len(vx_1)):
    plt.scatter(vx_1[i], vy_1[i])
for i in range(len(vx_2)):
    plt.scatter(vx_2[i], vy_2[i])
for i in range(len(vx_3)):
    plt.scatter(vx_3[i], vy_3[i])


#IF I WANT THE PLOT OF THE ORBITS
plt.plot(vx_1, vy_1, label='1', color = 'red')
plt.plot(vx_2, vy_2, label='2', color = 'green')
plt.plot(vx_3, vy_3, label='3', color = 'blue')

plt.title('three masses simulation...')
plt.xlabel('vx')
plt.ylabel('vy')
#plt.gca().set_aspect('equal', adjustable='box')
plt.grid(True)
plt.legend()
plt.show()
```
